## 선 그래프

### 기본적인 선 그래프 그리기

* 2차원 선 그래프 : plt.plot([x,]y[,fmt])

* 인수 생략 가능한 경우
    * plt.plot(y)
    * plt.plot(y, fmt)
    * plt.plot(x, y)
    * plt.plot(x, y, fmt)
 

```python
data1 = [10, 14, 19, 20, 25]
```

```python
plt.plot(data1) # x값은 지정 안했기에 y의 개수만큼 1씩 증가
```

```python
plt.plot(data1)
plt.show() # <- 그래프의 객체 정보가 출력이 안된다.
```

* x값과 y값 모두 있는 데이터로 선 그래프 그리기(y=2x^2)

```python
# numpy 모듈 호출
import numpy as np

x = np.arange(-4.5, 5, 0.5) # 배열 x를 생성. 범위 : [-4.5, 5], 0.5씩 증가
y = 2*x**2 # 수식을 이용해 배열 x에 대응하는 배열 y 생성
[x,y]
```

```python
plt.plot(x,y)
plt.show()
```

### 여러 그래프 그리기

```python
import numpy as np

x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10

import matplotlib.pyplot as plt
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
# 한 문장에서 같이 호출
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
# 따로 나오게 하고 싶을때
plt.plot(x, y1)
plt.figure() # 새로운 창 생성!
plt.plot(x, y2)
plt.figure(0)
plt.plot(x, y3)
plt.show()
```

```python
import numpy as np

# 데이터 생성
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2 # NumPy에서 cos()는 np.cos()으로 입력

plt.figure(1) # 1번 그래프 창을 생성
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그린다.

plt.figure(2) # 2번 그래프 창을 생성
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(1) # 이미 생성된 1번 그래프 창을 지정
plt.plot(x, y2)  # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 이미 생성된 2번 그래프 창을 지정
plt.clf() # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.show()
```
